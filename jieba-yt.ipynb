```python
import jieba
document = ["我喜歡寫程式", "我今年六月中會回台灣"]

for s in document:
    seg_list = jieba.cut(s)
    print('精準模式')
    print('/'.join(seg_list))
print('---------------')
for s in document:
    seg_list = jieba.cut(s, cut_all = True)
    print('全模式')
    print('/'.join(seg_list))
print('---------------')
for s in document:
    seg_list = jieba.cut_for_search(s)
    print('搜尋引擎')
    print('/'.join(seg_list))
```

```python
import jieba
document = ["我喜歡寫程式", "我今年六月中會回台灣"]

for s in document:
    seg_list = jieba.lcut(s)
    print('精準模式')
    print(seg_list)
print('---------------')
for s in document:
    seg_list = jieba.cut(s, cut_all = True)
    print('全模式')
    print('/'.join(seg_list))
print('---------------')
for s in document:
    seg_list = jieba.cut_for_search(s)
    print('搜尋引擎')
    print('/'.join(seg_list))
```

```python
import csv
import jieba
import pandas as pd

file = open('C:\\Users\\yisha\\Youtube_crawl\\77yt_titles.csv').read().split('\n')
keyword_cnt = {} 

for i in range(len(file)):
    result = []
    seg_list = jieba.cut(file[i])
    for w in seg_list:
        if w not in keyword_cnt:
            result.append(w)
            keyword_cnt[w] = 0
        keyword_cnt[w] += 1

outcome = open('C:\\Users\\yisha\\Youtube_crawl\\77yt_jieba_result.csv', 'w')
writer = csv.writer(outcome)
writer.writerows(keyword_cnt)

outcome.close()
```

```python
import jieba
import pandas as pd

file = open('C:\\Users\\yisha\\Youtube_crawl\\77yt_titles2.csv').read().split('\n')
keyword_cnt = {} 

for i in range(len(file)):
    seg_list = jieba.cut(file[i])
    for w in seg_list:
        if w not in keyword_cnt:
            keyword_cnt[w] = 0
        keyword_cnt[w] += 1
print(keyword_cnt)
```
